```python
import pandas as pd
import plotly.express as px
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline

print("Setup Complete")
```

```python
df = pd.read_csv('avocado-updated-2020.csv')
df.info()
```

```python
df.head()
```

```python
print(df['type'].value_counts(dropna=False))
print(df['geography'].value_counts(dropna=False))
```

```python
#Create line chart for Atlanta
msk = df['geography'] == 'Atlanta'

px.line(df[msk], x='date', y='average_price', color='type')
```

```python
#We can see that the organic avocado is more expensive in avarage than the conventional one in Atlanta
```

```python
#Create scatter plot 
fig = px.scatter(df, x='total_volume', y='average_price')
fig.show()
```

```python
# We can see that the when more avocados are produce (+volume) the average price decrease
```
